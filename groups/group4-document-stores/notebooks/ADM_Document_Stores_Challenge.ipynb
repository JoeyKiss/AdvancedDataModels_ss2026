## Hilfsfunktion zur schöneren Darstellung der Ergebnisse

Einfach ausführen!


````python
def show(cursor_or_list, title=""):
    results = list(cursor_or_list)
    if title:
        print(f"\n{'='*62}")
        print(f"  {title}  ({len(results)} Ergebnisse)")
        print(f"{'='*62}")
    if not results:
        print("→ Keine Ergebnisse gefunden.")
        return
    def flatten(doc, prefix=""):
        flat = {}
        for k, v in doc.items():
            key = f"{prefix}{k}"
            if isinstance(v, dict):
                flat.update(flatten(v, prefix=f"{key}."))
            elif isinstance(v, list):
                flat[key] = ", ".join(str(i) for i in v)
            else:
                flat[key] = v
        return flat
    flat_results = [flatten(r) for r in results]
    print(tabulate(flat_results, headers="keys", tablefmt="rounded_outline",
                   floatfmt=".2f", numalign="right"))
````
